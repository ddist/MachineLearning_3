## 1) Small circle inside Large circle

```python
import numpy as np
from sklearn.utils import check_random_state
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
def do_circles(n=2000,noisy_n=0.05):
    generator = check_random_state(10)
    linspace = np.linspace(0, 2 * np.pi, n // 2 + 1)[:-1]
    outer_circ_x = np.cos(linspace)
    outer_circ_y = np.sin(linspace)
    inner_circ_x = outer_circ_x * .3
    inner_circ_y = outer_circ_y * .3
    X = np.vstack((np.append(outer_circ_x, inner_circ_x), np.append(outer_circ_y, inner_circ_y))).T
    y = np.hstack([np.zeros(n // 2, dtype=np.intp), np.ones(n // 2, dtype=np.intp)])
    X += generator.normal(scale=noisy_n, size=X.shape)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    return X_train,y_train,X_test,y_test
```

```python
X_train, y_train, X_test, y_test = do_circles()
fig, axes = plt.subplots(1,2)
fig.set_size_inches(16,6)
axes[0].plot(X_train[y_train==0][:,0],X_train[y_train==0][:,1], '.g')
axes[0].plot(X_train[y_train==1][:,0],X_train[y_train==1][:,1], '.b')
axes[0].axis("equal")
axes[0].set_ylabel("y")
axes[0].set_xlabel("x")
axes[0].set_title("Training Set - N=" + str(1000) + "  n=" + str(0.05))
axes[1].plot(X_test[y_test==0][:,0],X_test[y_test==0][:,1], '.g')
axes[1].plot(X_test[y_test==1][:,0],X_test[y_test==1][:,1], '.b')
axes[1].axis("equal")
axes[1].set_ylabel("y")
axes[1].set_xlabel("x")
axes[1].set_title("Testing Set - N=" + str(1000) + "  n=" + str(0.05))
plt.show()
```

```python
def plot_classifier(clf,X_train,Y_train,X_test,Y_test,model_type, levels=None):
    x_min, x_max = np.minimum(X_train[:,0].min(), X_test[:,0].min()) - 0.2, np.maximum(X_train[:,0].max(), X_test[:,0].max()) + 0.2
    y_min, y_max = np.minimum(X_train[:,1].min(), X_test[:,1].min()) - 0.2, np.maximum(X_train[:,1].max(), X_test[:,1].max()) + 0.2
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
    
    f, axes = plt.subplots(1, 2, sharex="col", sharey="row",figsize=(16, 6))
    if model_type=="ann":
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    elif model_type=="tree":
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,0]
    else:
        raise ValueError("Model type not supported")
    
    Z = Z.reshape(xx.shape)
    axes[0].contourf(xx, yy, Z, alpha=0.6, cmap="GnBu", levels=levels)
    axes[1].contourf(xx, yy, Z, alpha=0.6, cmap="GnBu", levels=levels)
    axes[0].scatter(X_train[Y_train==0][:,0], X_train[Y_train==0][:,1], c='g', s=20, edgecolor='k', lw=0.5)
    axes[0].scatter(X_train[Y_train==1][:,0], X_train[Y_train==1][:,1], c='b', s=20, edgecolor='k', lw=0.5)
    axes[1].scatter(X_test[Y_test==0][:,0], X_test[Y_test==0][:,1], c='g', s=20, edgecolor='k', lw=0.5)
    axes[1].scatter(X_test[Y_test==1][:,0], X_test[Y_test==1][:,1], c='b', s=20, edgecolor='k', lw=0.5)
    axes[0].axis("equal")
    axes[1].axis("equal")
    axes[0].set_title("Training Set")
    axes[1].set_title("Test Set")
    plt.show()
```

```python
np.random.seed(42)
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, RMSprop, Adagrad
from keras.regularizers import l1, l2
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

def create_model(units=1, init="uniform", lr=1, activation="sigmoid", optimizer="sgd", regularizer=None, lam=0.01):
    model = Sequential()
    if regularizer=="l1":
        reg = l1(lam)
    elif regularizer=="l2":
        reg = l2(lam)
    else:
        reg = None
    model.add(Dense(units, input_dim=2, kernel_initializer=init, activation="relu", kernel_regularizer=reg))
    model.add(Dense(1, kernel_initializer=init, activation=activation))
    if optimizer=="sgd":
        opt = SGD(lr=lr)
    elif optimizer=="rmsprop":
        opt = RMSprop(lr=lr)
    else: 
        opt = Adagrad(lr=lr)
    model.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
    return model
```

```python
model = create_model()
model.fit(X_train, y_train, epochs=50, batch_size=100, verbose=0)
scores = model.evaluate(X_test, y_test, verbose=0)
print("Test Set Accuracy ANN 1 Unit: ", scores[1])
```

```python
plot_classifier(model, X_train, y_train, X_test, y_test, "ann", [0,0.5,1.0])
```

```python
from ipywidgets import interact, interact_manual, fixed
import ipywidgets as widgets

@interact(optimizer=["sgd", "rmsprop", "adagrad"], lr=[0.01,0.1,1], 
                 init=["uniform", "glorot_uniform", "ones"], regularizer=[None,"l1","l2"], lam=[0.001,0.01,0.1])
def interact_ann_1(units=fixed(1), optimizer="sgd", lr=1, init="uniform", regularizer=None, lam=0.1):
    model_ = create_model(units=units, init="glorot_uniform", optimizer=optimizer, lr=lr, regularizer=regularizer,
                         lam=lam)
    model_.fit(X_train, y_train, epochs=100, batch_size=100, verbose=0, shuffle=False)
    plot_classifier(model_, X_train, y_train, X_test, y_test, "ann")
    print("Test Set Accuracy = ", model_.evaluate(X_test, y_test, verbose=0)[1])
```

```python
# NO EJECUTAR, SE DEMORA 20+ MINS EN UN I7 4/8
clf = KerasClassifier(build_fn=create_model, verbose=0)
optimizers = ["sgd", "rmsprop", "adagrad"]
lrs = [0.01, 0.1, 1]
inits = ["uniform", "glorot_uniform", "ones"]
regularizers = [None, "l1", "l2"]
lams = [0.01, 0.1]
epochs = [30, 50]
activations = ["sigmoid", "tanh"]
param_grid = dict(optimizer=optimizers,lr=lrs,init=inits,regularizer=regularizers,
                  lam=lams,activation=activations,epochs=epochs)

grid = GridSearchCV(estimator=clf, param_grid=param_grid, n_jobs=4)
grid_result = grid.fit(X_train, y_train)
```

```python
# SI LA CELDA ANTERIOR NO SE EJECUTO, TAMPOCO EJECUTAR
for (k,v) in grid_result.best_params_.items():
    if k=="lam" and grid_result.best_params_["regularizer"]==None:
        continue
    print("{:<20}".format(k), v)
```
