## 3) Calidad de un Vino

En esta sección se estudia el problema de clasificar un vino como de buena o mala calidad, considerando algunos de sus atributos químicos. Para lograrlo, se experimenta con árboles de clasificación y técnicas genéricas de ensamblado de máquinas.

### a) Dataset

En esta sección se carga el conjunto de datos, que se encuentra separado en datos de vino tinto y blanco. Ámbos se unen y se agrega un nuevo atributo, que indica si el vino es tinto o blanco.

```python
import pandas as pd
import numpy as np
%matplotlib inline
np.random.seed(42)

df_red = pd.read_csv("winequality-red.csv",sep=";")
df_red["type"] = pd.Series(np.zeros(len(df_red.index)))
df_white = pd.read_csv("winequality-white.csv",sep=";")
df_white["type"] = pd.Series(np.ones(len(df_white.index)))
df = pd.concat([df_red,df_white], axis=0, ignore_index=True)
```

```python
df.describe()
```

Se observa que los datos tienen 12 atributos, correspondientes a características químicas de cada vino. También, hay mas datos relacionados a vino blanco que tinto. Los datos no están normalizados.

### b) Umbral de calidad

Como se desea abordar el problema como uno de clasificación, se considera un umbral de calidad, donde aquellos vinos con calidad mayor a 5 se consideran buenos, y malos en caso contrario. De esta forma se obtiene un problema de clasificación binaria.

```python
df["good_quality"] = [1 if q>5 else 0 for q in df.quality]
df = df.drop("quality", axis=1)
```

```python
h = df.hist(figsize=(16,10))
```

La cantidad de atributos se mantiene, pero se cambia la etiqueta por una binaria. Notar que las clases se encuentran desbalanceadas, donde la cantidad de vinos de buena calidad son aproximadamente el doble que los de mala calidad. 

Finalmente se divide en conjuntos de entrenamiento y pruebas, apartando un 25% de los datos para éste último.

```python
from sklearn.model_selection import train_test_split

x_tr, x_te, y_tr, y_te = train_test_split(df[df.columns[:df.columns.size-1]], df[df.columns[-1]], test_size=0.25)
print(x_tr.shape, y_tr.shape, x_te.shape, y_te.shape)
```

### c) Árbol de clasificación multinivel

El primer modelo a utilizar es el árbol de decisión como clasificador, donde los hiper-parámetros óptimos se encuentran con *K-Fold Cross Validation*. El puntaje para medir desempeño es el F1 Score, debido al desbalance de las clases.

En adelante se usará la siguiente función para graficar las matrices de confusión.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV

splitters = ["best", "random"]
criterions = ["gini", "entropy"]
max_depths = [None,5,15,20,30,40]

param_grid = dict(  criterion=criterions,
                    splitter=splitters,
                    max_depth=max_depths)

grid = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param_grid, scoring="f1")
results = grid.fit(x_tr, y_tr)
```

```python
results.best_params_
```

```python
results.best_score_
```

```python
classes = ["Bad", "Good"]
plot_confusion_matrix(confusion_matrix(y_te, grid.predict(x_te)), classes)
```

El puntaje para los mejores hiper-parámetros es de 0.81. La matriz de confusión indica que al algoritmo tiene problemas en clasificar la clase de vinos de mala calidad, con cerca de un 50% de clasificación errónea para esa clase.

### d) Random Forest

El ensamblador Random Forest entrena múltiples árboles de decisión, sobre un subconjunto aleatorio de atributos, para luego promediar la salida de los estimadores. Como vimos en la sección anterior, un único árbol de clasificación tiene problemas para etiquetar los vinos de mala calidad, probablemente debido al desbalance de las clases. Se espera que este ensamblado sea capaz de atenuar el efecto negativo del desbalance, y mejore el puntaje final del estimador.

A continuación se grafica la F1 Score sobre la cantidad de estimadores que RandomForest entrena. Además, se incluye la matriz de confusión del mejor estimador encontrado. Se experimentará con árboles poco profundos y profundos, para ver como se comporta en ensamblado en cada caso.

#### Max Depth = 5

```python
from sklearn.ensemble import RandomForestClassifier

scores = []
best_score = 0
best_clf = None
for n in range(1,102,10):
    model = RandomForestClassifier(n_estimators=n, max_depth=5, n_jobs=-1)
    model.fit(x_tr, y_tr)
    y_hat_tr = model.predict(x_tr)
    y_hat_te = model.predict(x_te)
    s_tr = f1_score(y_tr, y_hat_tr)
    s_te= f1_score(y_te, y_hat_te)
    scores.append([s_tr, s_te])
    if s_te > best_score:
        best_clf = model
scores = np.array(scores)
```

```python
f, ax = plt.subplots(1)
f.set_size_inches(12,6)
ax.plot(np.arange(1,102,10), scores[:,0], label="Train error")
ax.plot(np.arange(1,102,10), scores[:,1], label="Test error")
ax.set_ylim([0.5,1])
ax.set_xticks(np.arange(0,102,10))
ax.set_ylabel("F1 Score")
ax.set_xlabel("Number of Trees")
ax.set_title("F1 Score for n estimators Random Forest, with tree max depth = 5")
ax.legend()
plt.show()
print("Test max F1 score = ", scores[:,1].max())
plot_confusion_matrix(confusion_matrix(y_te, best_clf.predict(x_te)), classes)
```

#### Max Depth = 15

```python
scores = []
best_score = 0
best_clf = None
for n in range(1,102,10):
    model = RandomForestClassifier(n_estimators=n, max_depth=15, n_jobs=-1)
    model.fit(x_tr, y_tr)
    y_hat_tr = model.predict(x_tr)
    y_hat_te = model.predict(x_te)
    s_tr = f1_score(y_tr, y_hat_tr)
    s_te= f1_score(y_te, y_hat_te)
    scores.append([s_tr, s_te])
    if s_te > best_score:
        best_clf = model
scores = np.array(scores)
```

```python
f, ax = plt.subplots(1)
f.set_size_inches(12,6)
ax.plot(np.arange(1,102,10), scores[:,0], label="Train error")
ax.plot(np.arange(1,102,10), scores[:,1], label="Test error")
ax.set_ylim([0.5,1])
ax.set_xticks(np.arange(0,102,10))
ax.set_ylabel("F1 Score")
ax.set_xlabel("Number of Trees")
ax.set_title("F1 Score for n estimators Random Forest, with tree max depth = 15")
ax.legend()
plt.show()
print("Test max F1 score = ", scores[:,1].max())
plot_confusion_matrix(confusion_matrix(y_te, best_clf.predict(x_te)), classes)
```

Se observa que el ensamblado de árboles poco profundos logra el mismo resultado que un solo árbol profundo, y por la tendencia del gráfico, no parece posible que agregar más estimadores mejore el resultado.

El ensamblado de árboles profundos en cambio, si logró una mejora, con un puntaje de 0.86. 

### e) AdaBoost

De igual forma que en la sección anterior, se entrenará un ensamblado de árboles de decisión, pero esta vez utilizando AdaBoost, donde en cada estimador débil subsecuente son ajustados para tratar con aquellos datos de mayor dificultad para clasificar. Los estimadores pueden ser débiles, mientras su rendimiento sea levemente mayor a elección aleatoria (50%), puede probarse que la máquina de aprendizaje final converge a un estimador fuerte.

#### Max Depth = 5

```python
from sklearn.ensemble import AdaBoostClassifier

scores = []
best_score = 0
best_clf = None
for n in range(1,102,10):
    tree = DecisionTreeClassifier(max_depth=5)
    model = AdaBoostClassifier(base_estimator=tree, n_estimators=n)
    model.fit(x_tr, y_tr)
    y_hat_tr = model.predict(x_tr)
    y_hat_te = model.predict(x_te)
    s_tr = f1_score(y_tr, y_hat_tr)
    s_te= f1_score(y_te, y_hat_te)
    scores.append([s_tr, s_te])
    if s_te > best_score:
        best_clf = model
scores = np.array(scores)
```

```python
f, ax = plt.subplots(1)
f.set_size_inches(12,6)
ax.plot(np.arange(1,102,10), scores[:,0], label="Train error")
ax.plot(np.arange(1,102,10), scores[:,1], label="Test error")
ax.set_ylim([0.5,1])
ax.set_xticks(np.arange(0,102,10))
ax.set_ylabel("F1 Score")
ax.set_xlabel("Number of Trees")
ax.set_title("F1 Score for n estimators AdaBoost, with tree max depth = 5")
ax.legend()
plt.show()
print("Test max F1 score = ", scores[:,1].max())
plot_confusion_matrix(confusion_matrix(y_te, best_clf.predict(x_te)), classes)
```

#### Max Depth = 15

```python
from sklearn.ensemble import AdaBoostClassifier

scores = []
best_score = 0
best_clf = None
for n in range(1,102,10):
    tree = DecisionTreeClassifier(max_depth=15)
    model = AdaBoostClassifier(base_estimator=tree, n_estimators=n)
    model.fit(x_tr, y_tr)
    y_hat_tr = model.predict(x_tr)
    y_hat_te = model.predict(x_te)
    s_tr = f1_score(y_tr, y_hat_tr)
    s_te= f1_score(y_te, y_hat_te)
    scores.append([s_tr, s_te])
    if s_te > best_score:
        best_clf = model
scores = np.array(scores)
```

```python
f, ax = plt.subplots(1)
f.set_size_inches(12,6)
ax.plot(np.arange(1,102,10), scores[:,0], label="Train error")
ax.plot(np.arange(1,102,10), scores[:,1], label="Test error")
ax.set_ylim([0.5,1.1])
ax.set_xticks(np.arange(0,102,10))
ax.set_ylabel("F1 Score")
ax.set_xlabel("Number of Trees")
ax.set_title("F1 Score for n estimators AdaBoost, with tree max depth = 15")
ax.legend()
plt.show()
print("Test max F1 score = ", scores[:,1].max())
plot_confusion_matrix(confusion_matrix(y_te, best_clf.predict(x_te)), classes)
```

Esta vez si se observan mejoras para los árboles poco profundos, a diferencia de RandomForest, lo que puede explicarse por la capacidad de AdaBoost de trabajar con estimadores débiles, convergiendo a uno fuerte. El resultado es similar para árboles profundos, no solo en el puntaje final, si no que en el error de clasificación para cada clase. 

Podría esperarse que AdaBoost logre mejores puntajes disminuyendo los errores de clasificación para las clases más dificiles, pero no es el caso en árboles profundos. Para los árboles poco profundos, si se observa que el número de falsos positivos y falsos negativos son casi iguales, lo que hace a ambas clases igual de dificiles de clasificar.

### f) SVM

Se propone usar SVM con kernel gaussiano para resolver el problema de la calidad del Vino. Los hiper-parámetros a usar son los indicados en el código, luego de una breve búsqueda manual.

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_tr)
x_tr_s = scaler.transform(x_tr)
x_te_s = scaler.transform(x_te)

clf = SVC(C=100, kernel="rbf", gamma=0.99)
clf.fit(x_tr_s, y_tr)
print("Training Set - F1 score = ", f1_score(y_tr, clf.predict(x_tr_s)))
print("Test Set - F1 score = ", f1_score(y_te, clf.predict(x_te_s)))
plot_confusion_matrix(confusion_matrix(y_te, clf.predict(x_te_s)), classes)
```

El resutado es mejor que el de un único árbol profundo pero peor que un ensamblado de árboles. De todas formas, no se puede considerar como un clasificador débil. A continuación se usa AdaBoost para entrenar un ensamblado de SVM.

#### AdaBoost + SVM

```python
scores = []
best_score = 0
best_clf = None
for n in range(1,102,10):
    clf = SVC(C=100, kernel="rbf", gamma=0.99, probability=True)
    model = AdaBoostClassifier(base_estimator=cfl, n_estimators=n)
    model.fit(x_tr_s, y_tr)
    y_hat_tr = model.predict(x_tr_s)
    y_hat_te = model.predict(x_te_s)
    s_tr = f1_score(y_tr, y_hat_tr)
    s_te= f1_score(y_te, y_hat_te)
    scores.append([s_tr, s_te])
    if s_te > best_score:
        best_clf = model
scores = np.array(scores)
```

```python
f, ax = plt.subplots(1)
f.set_size_inches(12,6)
ax.plot(np.arange(1,102,10), scores[:,0], label="Train error")
ax.plot(np.arange(1,102,10), scores[:,1], label="Test error")
ax.set_ylim([0.5,1.1])
ax.set_xticks(np.arange(0,102,10))
ax.set_ylabel("F1 Score")
ax.set_xlabel("Number of Trees")
ax.set_title("F1 Score for n estimators AdaBoost, with SVM kernel=rbf and C=100")
ax.legend()
plt.show()
print("Test max F1 score = ", scores[:,1].max())
plot_confusion_matrix(confusion_matrix(y_te, best_clf.predict(x_te)), classes)
```

Se observa que el ensamblado no aprendió a clasificar, ya que etiqueta todas los datos como vinos de buena calidad. Esto puede deberse a que la SVM no es un clasificador débil, por lo que no funciona como se espera con AdaBoost. Se podría debilitar deliberadamente las SVM, por ejemplo, entrenandolas con un subconjunto aleatorio del conjunto de entrenamiento, o usando Bagging en vez de AdaBoost, pero dichos experimentos no se realizarán en esta ocasión.

### g) Importancia de características

En un árbol de clasificación, la profundidad a la que se encuentra una característica, es decir, el nivel del nodo donde se considera para hacer una división (*split*) puede usarse como métrica de importancia, ya que las características en los niveles altos contribuyen a la predicción final de una mayor fracción del total de ejemplos. Si se promedia la importancia por cada árbol del Random Forest, se puede obtener la importancia a nivel de ensamblado. Esta métrica ya se encuentra implementada en la clase RandomForest de *sklearn*, por lo que a continuación se muestra el ranking de características para el ensamblado por RandomForest con árboles profundos entrenado en la sección d).

```python
print("Feature Importance Ranking\n")
print("{:<5} {:<25}".format("Rank", "Feature"), "Importance\n")
indices = np.flip(np.argsort(model.feature_importances_), axis=0)
i = 0
for v in model.feature_importances_[indices]:
    print("{:<5} {:<25}".format(i+1, df.columns[indices[i]]), "{:3.2f}%".format(v*100))
    i = i+1
```

Si el estimador base en un ensamblado permite calcular la importancia por el criterio descrito, como es el caso de un Árbol de Clasificación, entonces si es posible implementar en AdaBoost. De igual forma que con Random Forest, se promedia la importancia de cada árbol, obteniendo la importancia a nivel de ensamblado. Si el estimador base no es algún tipo de árbol, entonces no es posible implementar este criterio.
